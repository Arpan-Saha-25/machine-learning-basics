```python
marksUT= {'Name':['Raman','Raman','Raman','Zuhaire','Zuhaire','Zuhaire', 'Ashravy','Ashravy','Ashravy','Mishti','Mishti','Mishti'],
 'UT':[1,2,3,1,2,3,1,2,3,1,2,3],
 'Maths':[22,21,14,20,23,22,23,24,12,15,18,17],
 'Science':[21,20,19,17,15,18,19,22,25,22,21,18],
 'S.St':[18,17,15,22,21,19,20,24,19,25,25,20],
 'Hindi':[20,22,24,24,25,23,15,17,21,22,24,25],
 'Eng':[21,24,23,19,15,13,22,21,23,22,23,20]
 }

df = pd.DataFrame(marksUT)
df
```

```python
print(df.max())             # max values columns-wise
print(df.max(axis=0))       # max values columns-wise
print("\n", df.max(numeric_only=True))     # max values columns-wise
print("\n", df[df.UT==2])           # marks for each subject where UT=2 
print("\n", df[df.UT==2].max())     # max marks for each subject where UT=2 
```

```python
df_col_remove = df.drop("Name",axis=1)
print("\n---------------------max row-wise-----------\n", df_col_remove.max(axis=1))
print("\n---------------------max col-wise-----------\n", df_col_remove.max(axis=0))
```

```python
print("\n---------------------min row-wise-----------\n", df_col_remove.min(axis=1))
```

```python
df_mishti = df.loc[df.Name=="Mishti"]
print("\n", df_mishti)
print("\n", df_mishti.min())
print("\n", df_mishti[["Maths","Science"]].min())
print("\n", df_mishti[["Maths","Science"]].max())
print("\n", df_mishti[["Maths","Science"]].sum())
```

```python
print("\n", df_mishti[["Maths","Science"]].count())
print("\n", df_mishti[["Maths","Science"]].count(axis=1))
```

```python
print(df_mishti[["Maths"]].mean())
print(df_mishti[["Maths"]].median())
print(df_mishti[["Maths"]].mode())
```

```python
print(df_mishti[["Maths"]].quantile())    # returns the value below which a given percentage (quantile) of the data falls.
```

```python
print(df_mishti[["Maths"]].std()) # std deviation
print(df_mishti[["Maths"]].var())   # variance
print(df_mishti[["Maths"]].describe()) # sumarises all stats
```
