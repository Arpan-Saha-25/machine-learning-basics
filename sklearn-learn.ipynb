```python
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[2600], [3000], [3200], [3600], [4000]])
y = np.array([540000, 610000, 720000, 860000, 910000])
```

```python
plt.xlabel("Area (sq feets)")
plt.ylabel("Price in USD")
# plt.scatter(X,y,color="red", marker="+")
```

```python
reg = LinearRegression()

# * model.fit() - trains the model by fitting data into it. (X should always be 2D)
reg.fit(X,y)        
```

```python
reg.predict([[3400]])   # X should be 2D HERE also
```

```python
# to find the coefficient or slope m 
# eqn : (y = mx + b)
m = reg.coef_
m
```

```python
# intercept (b)
b = reg.intercept_
b
```

```python
newInput = np.array([[1230],[2100],[3000],[4100], [5000]])
predicted = reg.predict(newInput)
predicted
```

```python
plt.plot(X,reg.predict(X),color='blue')
plt.scatter(X,y,color="red", marker="+")
plt.scatter(X,reg.predict(newInput), color="black", marker="*")
```
