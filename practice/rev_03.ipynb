## Feature Scaling 
- Method to scale numeric features in the same scale or range
- Last step in data preprocessing.
- aka data normalization.
- We apply FS on independent variables.

- If height is 140 cm and 8.2 feet. Here, we use FS.

- KNN, K MEANS, SVM , PCA -> We have to scale features as for calculating distances and vectors.

- Tree based algo like decision tree, random forest etc doesnt need fs.

#### STANDARDIZATION :
- RESCALE THE FEATURE SUCH AS MEAN=0 AND SD=1.
- Z-SCORE = ```(X - MEAN ) / SD```
- IF DATA FOLLOW `NORMAL DISTRIBUTION` THEN WE CAN APPLY THIS BUT SHAPE WILL REMAIN INTACT.

#### NORMALISATION
- WE SCALE THE FEATURE IN FIXED RANGE BETWEEN 0 - 1.
- X_NORM = `X - X_MIN / X_MAX - X_MIN`
- AKA `MIN-MAX SCALING` .

THERE IS NO THUMB RULE TO USE STANDARDISATION OR NORMALISATION FOR SPECIAL ML ALGO.

````python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
````

````python
df = sns.load_dataset('titanic')
df
````

````python
df_main = df[['survived','pclass','age','parch']]
df_main.head()
````

````python
# to get rid of nan values 
df3 = df_main.fillna(df_main.mean())
df3.head()
````

````python
X = df3.drop("survived", axis=1)
Y = df3["survived"]
X.shape, Y.shape
````

````python
x_train, x_test, y_train, y_test =  train_test_split(X, Y, test_size=0.2, random_state=51)
````

````python
# x_train.head() 
# y_train.tail()
````

````python
sc = StandardScaler()
sc.fit(x_train)
sc.mean_ 
sc.scale_
````

````python
x_train_sc = sc.transform(x_train)
x_test_sc = sc.transform(x_test)
````

````python
x_train_sc = pd.DataFrame(x_train_sc, columns=["pclass","age","parch"])
x_test_sc = pd.DataFrame(x_test_sc, columns=["pclass","age","parch"])
````
