### Matplotlib for Python

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
x = np.linspace(1,100,10)
y = np.random.random(10) * 10 
plt.plot(x,y)
plt.show()
```

```python
plt.title("Revision of Matplotlib")
plt.plot(x,y)
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.show()
```

```python
x2 = np.linspace(1,100,10)
y2 = np.random.random(10) * 10 

plt.plot(x,y , label = "line1" , linewidth = 6)
plt.plot(x2,y2 , label = "line2")
plt.legend()
plt.show()
```
