```python
date1 = ["30/4" , "1/5", "2/5", "3/5", "4/5"]
# date1 = ["30/4" , "1/5", "2/5"]
temp1 = [27, 10, 31, 34, 45]
temp2 = np.random.randint(10,50,size=5)


plt.grid(True)

plt.xlabel("Date")
plt.ylabel("Temp")


plt.plot(date1,temp1, color="red", marker=7)
plt.plot(date1,temp2, color="green", marker=6)
# plt.savefig("p11_date_temp")    # saved as png

plt.legend(["black","purple"], loc = "right")
```

```python
import pandas as pd

file = pd.read_csv("data_meal.csv")
df = pd.DataFrame(file)
# file.plot()
# pd.plot(df)
# df
```

```python
plt.plot(df["Day"], df["Week 1"], marker="x",markersize = 10)
plt.plot(df["Day"], df["Week 2"], marker="o",markersize = 10,linestyle='dashed')
plt.plot(df["Day"], df["Week 3"], marker="o",markersize = 10,linewidth=3,linestyle="dotted")
plt.xticks(rotation=30)
plt.yticks(rotation=30)
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
plt.grid(color='lightgreen')
# plt.show()
```
